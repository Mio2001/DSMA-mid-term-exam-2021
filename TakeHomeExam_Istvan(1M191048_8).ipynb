```python
import pandas as pd
from google.colab import drive
import numpy as np 
drive.mount('/content/drive')
```

```python
#1 and partly 2?
#read in that file as a dataframe, set that index
ramen = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ramen-ratings.csv', index_col = 0)
```

```python
#the other part of 2?
#rename and print out overview
ramenR = ramen.rename(columns={"Top Ten": "Top", "Stars": "Score"})
ramenR
```

```python
#3
#subset of the dataframe that only has the information of the brand of ramen and its score
#not groupby, or creation of new dataframe, access specific rows or columns of data)
ramenR[["Brand","Score"]]
```

```python
#4 convert null values to NaN...it looks like it already IDK
#ITS WHITE SPACE RN THATS WHY
#ITS UNDER TOP YOU IDIOT
ramenR["Top"].fillna("NaN", inplace = True)
ramenR
```

```python
ramenR.isnull().sum()
#did we get rid of all the null values in the last column?
```

```python
print(ramenR.dtypes)
#aight why ain't number 5 working out...
#the score values aren't floats
```

```python
#change them objects to floats
ramenR['Score'] = pd.to_numeric(ramenR['Score'], errors='coerce')
```

```python
print(ramenR.Score.dtypes)
#LETS GOOOOO
```

```python
#5 Nissin, rating is above 3.5
#ramenR[ramenR['Nissin'].isin(Brand)]
#the above results in an error
#highscore = ramenR[ramenR['Score'] > 3.5]
#the above only results in the scores above 3.5
#(trumpinsults.insult == "dumb").sum()
#(ramenR.Brand == "Nissin").sum()
#trumpinsults.loc[(trumpinsults.insult == "dumb")]
ramenR.loc[(ramenR.Brand == "Nissin") & (ramenR.Score > 3.5)]
```

```python
##6 Group the ramen by brand and calculate the mean, max, and min values in a single output.
brand = ramenR.groupby('Brand').Score.agg([len, min, max,np.std,np.mean])
brand
#yay!
```

```python
#7 What is the mode rating for all ramen?
ramenR.Score.mode()
#it appears to be 4.0
```

```python
#8 Groupby brand and variety, display descending order (highest to lowest)
brand_variety = ramenR.groupby(['Brand', 'Variety']).Score.agg([np.mean])
brand_variety
```

```python
#test = df.sort_values(['one'], ascending=[False])
brandvariety_down = brand_variety.sort_values(['mean'], ascending=[False])
brandvariety_down
```

```python
#highscore = ramenR[ramenR['Score'] > 3.5]
highest_average = brand_variety[brand_variety['mean'] == 5.0]
highest_average
```
